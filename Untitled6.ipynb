```python
def check_input(s):
    if type(s) == str:
        return True
    else:
        return False

def get_substring(s):
    unique = ''
    max_unique = ''
    i = 0
    first = 0
    start = True
    while i < len(s):
        if start:
            unique += s[i]
            i += 1
            start = False
        else:
            if s[i] in unique:
                if len(unique) > len(max_unique):
                    max_unique = unique
                first += 1
                unique = s[first]
                i = first + 1
            else:
              unique += s[i]
              i += 1
    if len(unique) > len(max_unique):
        max_unique = unique
    return max_unique
def main(lst):
    max_s = ''
    ok = False
    for s in lst:
        if check_input(s):
            ok = True
            if len(s) > len(max_s):
              max_s = s
    if ok:
        print(str(lst.index(max_s)) + ", '" + get_substring(max_s) + "'")
    else:
        print("-1, ''")
```
